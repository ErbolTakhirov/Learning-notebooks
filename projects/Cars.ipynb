````python
def extract_kw(text):
    if not isinstance(text, str):
        return None
    import re
    match = re.search(r'\(?\s*(\d+)\s*[кk][вv][тt]?\s*\)?', text, flags=re.IGNORECASE)
    if match:
        return int(match.group(1))
    return None
````

````python
def fill_metric_by_first_word(df, metric, metric2, metric3=None):
    df = df.copy()
    name_to_body = {}
    for _, row in df.dropna(subset=[metric]).iterrows():
        first_word = row[metric2].split()[0]
        for __, row2 in df.dropna(subset=[metric]).iterrows():
          first_ = row[metric3].split()[0]
          if first_word not in name_to_body:
              name_to_body[[first_word, first_]] = row[metric]

    result = []
    if metric3 is not None:
        fill_metric_by_first_word(df, metric, metric3)


    def get_body_type(name):
        first_word = name.split()[0]
        return name_to_body.get(first_word, np.nan)

    mask = df[metric].isna()
    df.loc[mask, metric] = df.loc[mask, metric2].apply(get_body_type)

    return df
````

````python
def fill_metric_by_first_word(df, metric, metric2, metric3=None):
    df = df.copy()

    # Словарь для хранения значений: ключ — (first_word_metric2, first_word_metric3) или (first_word_metric2,)
    name_to_value = {}

    # Подготовим данные: только строки, где `metric` не NaN
    valid_rows = df.dropna(subset=[metric])

    for _, row in valid_rows.iterrows():
        # Получаем первое слово из metric2
        first_word_2 = row[metric2].strip().split()[0] if isinstance(row[metric2], str) else str(row[metric2]).split()[0]

        if metric3 is not None:
            # Если metric3 задан, используем и его первое слово
            first_word_3 = row[metric3].strip().split()[0] if isinstance(row[metric3], str) else str(row[metric3]).split()[0]
            key = (first_word_2, first_word_3)
        else:
            # Иначе ключ только по первому слову metric2
            key = (first_word_2,)

        # Сохраняем первое найденное значение для данного ключа
        if key not in name_to_value:
            name_to_value[key] = row[metric]

    # Функция для поиска значения по ключу
    def get_value(name, year=None):
        # Получаем первое слово из name
        first_word_2 = name.strip().split()[0] if isinstance(name, str) else str(name).split()[0]

        if metric3 is not None and year is not None:
            first_word_3 = year.strip().split()[0] if isinstance(year, str) else str(year).split()[0]
            key = (first_word_2, first_word_3)
        else:
            key = (first_word_2,)

        return name_to_value.get(key, np.nan)

    # Применяем заполнение только к строкам с NaN в `metric`
    mask = df[metric].isna()

    if metric3 is not None:
        # Используем оба столбца
        df.loc[mask, metric] = df.loc[mask].apply(
            lambda row: get_value(row[metric2], row[metric3]), axis=1
        )
    else:
        # Используем только metric2
        df.loc[mask, metric] = df.loc[mask][metric2].apply(get_value)

    return df
````

````python
# Шаг 1: Создаём словарь средних годов по condition
mean_year_by_cond = {}
for cond in cars['condition'].dropna().unique():
    mean_year = cars.loc[cars['condition'] == cond, 'production_year'].mean()
    if not pd.isna(mean_year):
        mean_year_by_cond[cond] = mean_year

# Шаг 2: Сортируем по годам
sorted_conds = sorted(mean_year_by_cond.items(), key=lambda x: x[1])
labels = [cond for cond, mean in sorted_conds]
bin_edges = [0]  # начальная граница

# Шаг 3: Добавляем середины между средними годами
for i in range(len(sorted_conds) - 1):
    mid = (sorted_conds[i][1] + sorted_conds[i+1][1]) / 2
    bin_edges.append(mid)
bin_edges.append(2030)  # верхняя граница — например, 2030 год

# Шаг 4: Печать результатов
print("Labels:", labels)
print("Bin edges:", bin_edges)
````

````python
def fill_by_nearest_year(row, df, metric):
    if pd.isna(row[metric]):
        known = df[df[metric].notna()]
        nearest_idx = (known['production_year'] - row['production_year']).abs().idxmin()
        return known.loc[nearest_idx, metric]
    else:
        return row[metric]
````
