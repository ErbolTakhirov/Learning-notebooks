Ранжирование

**Задание**

На основании колонки petal_length из датасета ирисов получите серию категориальных значений, определяющих принадлежность значения к одному из четырех интервалов между квартилями. Категориальные значения должны называться Q1, Q2, Q3, Q4.

Постройте скаттерплот, где по оси y будет отложен petal_length, а по х - индекс строки. С помощью параметра hue отобразите категории, созданные выше

```python
iris = sns.load_dataset('iris')
iris
```

```python
iris.describe()
```

```python
pd.cut(iris.sepal_length, bins=3, labels=['l1', 'l2', 'l3'])
```

```python
pd.cut(iris.sepal_length, bins=3, labels=['l1', 'l2', 'l3']).value_counts()
```

```python
pd.cut(iris.sepal_length, [0, 5, 7, 10], labels=['l1', 'l2', 'l3'])
```

```python
pd.cut(iris.sepal_length, [0, 5, 7, 10], labels=['l1', 'l2', 'l3']).value_counts()
```

```python
bins = np.concatenate([[0], iris.petal_length.quantile([0.25, 0.5, 0.75]).values, [iris.petal_length.max()]])
bins
```

```python
labels = ['Q1', 'Q2', 'Q3', 'Q4']
```

```python
iris['Q'] = pd.cut(iris.petal_length, bins=bins, labels=labels)
iris
```

```python
bins
```

```python
sns.scatterplot(data=iris, y='petal_length', x=iris.index, hue='Q')
```
