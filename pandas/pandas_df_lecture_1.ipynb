Арифметичсекие операции между датафреймом и numpy массивом.

image.png

Иногда бывает необходимость выполнить арифметическую операцию между датафреймом и одномерным массивом (серией). Данный одномерный массив можно поэлементно сложить (умножить, разделить...) с каждым элементом датафрейма ПОСТРОЧНО или ПО СТОЛБЦАМ. Поэтому общий вид даного функционала следующий:



```
df.add(array, axis=0/1, fill_value=0)
df.sub(array, axis=0/1, fill_value=0)
df.div(array, axis=0/1, fill_value=0)
df.mul(array, axis=0/1, fill_value=0)
df.pow(array, axis=0/1, fill_value=0)
```
Где
* df - датафрейм
* array - одномерный массив
* axis - по столбцам или по строкам
* Любая матоперация с NaN дает NaN. Поэтому есть возможность настроить заполнение пропусков любым указанным значением с помощью параметра `fill_value=...`
